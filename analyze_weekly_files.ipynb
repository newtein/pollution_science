```python
state_code_df = pd.read_csv("states_and_counties.csv", dtype = {'State Code': str, 'County Code': str})
state_code_df.head(2)
```

```python
state_lookup = {}
state_df = state_code_df[["State Code", "State Name"]]
for index, row in state_df.iterrows():
    state_code=row['State Code']
    state_name=row['State Name']
    state_lookup[state_code] = state_name
```

```python
county_lookup = {}
county_df = state_code_df[["County Code", "County Name"]]
for index, row in county_df.iterrows():
    county_code=row['County Code']
    county_name=row['County Name']
    county_lookup[county_code] = county_name
```

```python
df['state_name_1'] = df['state_code_1'].apply(lambda x: state_lookup.get(x))
df['state_name_2'] = df['state_code_2'].apply(lambda x: state_lookup.get(x))
```

```python
df['county_name_1'] = df['county_code_1'].apply(lambda x: county_lookup.get(x))
df['county_name_2'] = df['county_code_2'].apply(lambda x: county_lookup.get(x))
df
```
