```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fname = 'data/network_analysis_final_file.csv'
df = pd.read_csv(fname)
```

```python
plt.close()

for county in df['county'].unique():
    c_df = df[df['county']==county]
    x_ticks = c_df['date']
    x = range(len(c_df['date']))

    plt.plot(x, c_df['cases'], label=county)
plt.legend(loc='upper left')
plt.show()
```
