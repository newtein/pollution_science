```python
xticks = ['betweeness', 'closeness', 'clustering', 'communicability betweeness', 'current flow closeness', 
          'degree', 'harmonic', 'load']

f = open('SA_files/SA_5_12.csv', "r")
reader= csv.reader(f)
train, test = [], []
for row in reader:
    if row[2]:
        train.append(float(row[2]))
        test.append(float(row[3]))
f.close()

plt.close()

fig, ax = plt.subplots(figsize=(9, 5), dpi=600)
cmap= plt.get_cmap('GnBu')
x1 = range(0, len(train)*2, 2)
x2 = list(map(lambda x:x+0.5, x1))
x3 = list(map(lambda x:x+0.25, x1))

plt.bar(x1, train, color=cmap(0.99), width=0.5, label='Training data')
plt.bar(x2, test, color=cmap(0.5), width=0.5, label='Testing data')
plt.ylim(0, 1.2)
#plt.legend(loc='upper right')
plt.ylabel('R$^2$ score')
plt.title("Similarity Metric: Pearson correlation", size=17)
plt.xticks(x3, [i.replace(' ', '\n').capitalize() for i in xticks], rotation =0, size=8.5)
max_x, min_x = max(x2)+1, min(x1)-1
ax.hlines(1.0, min_x, max_x, color='r', linestyles='--', linewidth=0.5)

plt.xlim(min_x, max_x)
ax.text(-0.085, 1, 'a)', transform=ax.transAxes, size=17, color='green')


plt.savefig('manuscript_images/similarity_correlation_MR.png', bbox_inches='tight', dpi=600)
plt.show()
```

```python
xticks = ['betweeness', 'current flow betweeness', 'closeness', 'clustering', 'communicability betweeness', 'current flow closeness', 
          'degree', 'harmonic', 'load']
metrics = ['Euclidean distance', 'DTW', 'CDTW']
notes = ['b)', 'c)', 'd)']
for idx, filename in enumerate(['SA_13_21', 'SA_22_30', 'SA_31_39']):
    f = open('SA_files/{}.csv'.format(filename), "r")
    reader= csv.reader(f)
    train, test = [], []
    for row in reader:
        if row[2]:
            train.append(float(row[2]))
            test.append(float(row[3]))
    f.close()

    plt.close()

    fig, ax = plt.subplots(figsize=(9, 5), dpi=600)
    cmap= plt.get_cmap('GnBu')
    x1 = range(0, len(train)*2, 2)
    x2 = list(map(lambda x:x+0.5, x1))
    x3 = list(map(lambda x:x+0.25, x1))

    plt.bar(x1, train, color=cmap(0.99), width=0.5, label='Training data')
    plt.bar(x2, test, color=cmap(0.5), width=0.5, label='Testing data')
    plt.ylim(0, 1.2)

    plt.ylabel('R$^2$ score')
    plt.title("Similarity Metric: {}".format(metrics[idx]), size=17)
    plt.xticks(x3, [i.replace(' ', '\n').capitalize() for i in xticks], rotation =0, size=8.5)
    max_x, min_x = max(x2)+1, min(x1)-1
    ax.hlines(1.0, min_x, max_x, color='r', linestyles='--', linewidth=0.5)

    plt.xlim(min_x, max_x)
    ax.text(-0.085, 1, notes[idx], transform=ax.transAxes, size=17, color='green')
    
    if notes[idx] == 'b)':
        plt.legend(loc='upper right')


    plt.savefig('manuscript_images/similarity_{}_MR.png'.format(metrics[idx]), bbox_inches='tight', dpi=600)
    plt.show()
```
